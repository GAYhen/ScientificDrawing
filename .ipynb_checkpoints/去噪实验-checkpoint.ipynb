# 原噪声、加性噪声

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置信号参数
sampling_rate = 1000  # 采样率 Hz
duration = 1  # 信号持续时间（秒）

# 生成时间向量
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# 生成复杂信号（叠加不同频率和振幅的正弦波）
frequencies = [5, 10, 20]  # 频率列表
amplitudes = [1, 0.5, 0.8]  # 对应的振幅
signal = sum(amplitude * np.sin(2 * np.pi * frequency * t) for amplitude, frequency in zip(amplitudes, frequencies))

# 添加噪声
noise_amplitude = 0.5  # 增加噪声的振幅
noise = noise_amplitude * np.random.randn(len(t))  # 生成高密度噪声
noisy_signal = signal + noise  # 将噪声加到信号中

# 绘制两张子图
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# 绘制原始复杂信号
axs[0].plot(t, signal, label='Original Complex Signal', color='#8c96c6')
axs[0].set_title('Original Complex Signal')
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Amplitude')
axs[0].legend()
axs[0].grid(True)

# 绘制带噪声的信号
axs[1].plot(t, noisy_signal, label='Signal with More Noise', color='orange')
axs[1].set_title('Signal with More Noise')
axs[1].set_xlabel('Time [s]')
axs[1].set_ylabel('Amplitude')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

# AE编码器滤波

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# 设置字体和坐标轴
plt.rcParams['font.sans-serif'].insert(0, 'SimHei')
plt.rcParams['axes.unicode_minus'] = False

# 生成数据
np.random.seed(0)
t = np.linspace(0, 1, 1024)  # 修改采样点数为1024
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)
noise = np.random.normal(0, 0.5, signal.shape)
noisy_signal = signal + noise

# 归一化信号到[-1, 1]
signal_max = np.max(np.abs(signal))
signal_normalized = signal / signal_max
noisy_signal_normalized = noisy_signal / signal_max

# 定义自编码器
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv1d(1, 64, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2),
            nn.Conv1d(64, 32, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2),
            nn.Conv1d(32, 16, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2),
            nn.Conv1d(16, 8, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose1d(8, 16, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.ReLU(True),
            nn.ConvTranspose1d(16, 32, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.ReLU(True),
            nn.ConvTranspose1d(32, 64, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.ReLU(True),
            nn.ConvTranspose1d(64, 1, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.Tanh()  # 最终层使用Tanh保持输出在[-1, 1]之间
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# 初始化模型、损失函数和优化器
model = Autoencoder()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# 准备数据
X_train = torch.FloatTensor(noisy_signal_normalized).unsqueeze(0).unsqueeze(0)
Y_train = torch.FloatTensor(signal_normalized).unsqueeze(0).unsqueeze(0)

# 训练模型
num_epochs = 500
losses = []
for epoch in range(num_epochs):
    optimizer.zero_grad()
    output = model(X_train)
    loss = criterion(output, Y_train)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch+1}/{num_epochs} Loss: {loss.item():.4f}")

# 测试模型
model.eval()
with torch.no_grad():
    filtered_signal_normalized = model(X_train).squeeze(0).squeeze(0).numpy()
filtered_signal = filtered_signal_normalized * signal_max  # 恢复到原始幅度

# 计算性能指标
def compute_snr(signal, reference):
    noise = signal - reference
    snr = 10 * np.log10(np.sum(reference**2) / np.sum(noise**2))
    return snr

def compute_ncc(signal, reference):
    ncc = np.correlate(signal, reference, mode='full')
    return ncc[ncc.size // 2] / (np.sqrt(np.sum(signal**2) * np.sum(reference**2)))

snr_noisy = compute_snr(noisy_signal, signal)
snr_filtered = compute_snr(filtered_signal, signal)
ncc = compute_ncc(filtered_signal, signal)
mse = criterion(torch.FloatTensor(filtered_signal), torch.FloatTensor(signal)).item()
rmse = np.sqrt(mse)

print(f"SNR of Noisy Signal: {snr_noisy:.2f} dB")
print(f"SNR of Filtered Signal: {snr_filtered:.2f} dB")
print(f"NCC between Filtered and Original Signal: {ncc:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

# 绘制损失曲线和信号对比图
plt.figure(figsize=(10, 4))
plt.plot(losses, label='Training Loss')
plt.title("训练损失曲线")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 7))
plt.subplot(2,1,1)
plt.plot(t, signal, label='原始信号', color='green')
plt.plot(t, noisy_signal, label='噪声信号', color='red')
plt.legend()

plt.subplot(2,1,2)
plt.plot(t, signal, label='原始信号', color='green')
plt.plot(t, filtered_signal, label='自编码器滤波后的信号', color='blue')
plt.legend()
plt.show()
```

# 卡尔曼滤波

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成信号
np.random.seed(0)
t = np.linspace(0, 1, 1024)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)
measurement_noise = np.random.normal(0, 0.5, signal.shape)
noisy_signal = signal + measurement_noise

# 简化版卡尔曼滤波实现
n_iterations = len(t)
x_hat = np.zeros(n_iterations)

# 卡尔曼滤波的参数，也许需要根据实际应用进行调整
process_variance = 1e-5
estimated_error = 1.0
measurement_error = measurement_noise.var()

x_hat[0] = noisy_signal[0]

for k in range(1, n_iterations):
    kalman_gain = estimated_error / (estimated_error + measurement_error)
    x_hat[k] = x_hat[k - 1] + kalman_gain * (noisy_signal[k] - x_hat[k - 1])
    estimated_error = (1 - kalman_gain) * estimated_error + process_variance  # 更新估计的误差方差

# 计算NCC和SNR
def compute_snr(a, b):
    """计算信噪比(SNR)"""
    signal_power = np.mean(a**2)
    noise_power = np.mean((a - b)**2)
    snr = 10 * np.log10(signal_power / noise_power)
    return snr

def compute_ncc(a, b):
    """计算归一化互相关(NCC)"""
    ncc = np.correlate(a - np.mean(a), b - np.mean(b), mode='same')
    ncc /= np.sqrt(np.sum((a - np.mean(a)) ** 2) * np.sum((b - np.mean(b)) ** 2))
    ncc = np.max(ncc)  # 取最大值作为结果
    return ncc

snr_before = compute_snr(signal, noisy_signal)
snr_after = compute_snr(signal, x_hat)
ncc_value = compute_ncc(signal, x_hat)

print(f"SNR before filtering: {snr_before:.2f} dB")
print(f"SNR after filtering: {snr_after:.2f} dB")
print(f"NCC between filtered signal and original signal: {ncc_value:.4f}")

# 绘制对比图
plt.figure(figsize=(14, 7))
plt.subplot(2, 1, 1)
plt.plot(t, signal, label='原始信号', color='green')
plt.plot(t, noisy_signal, label='噪声信号', color='red')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, signal, label='原始信号', color='green')
plt.plot(t, x_hat, label='卡尔曼滤波后的信号', color='blue')
plt.legend()
plt.show()
```

# GAN+自编码器滤波（Conv）

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# 设置字体和坐标轴
plt.rcParams['font.sans-serif'].insert(0, 'SimHei')
plt.rcParams['axes.unicode_minus'] = False

# 生成数据
np.random.seed(0)
t = np.linspace(0, 1, 1024)  # 修改采样点数为1024
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)
noise = np.random.normal(0, 0.5, signal.shape)
noisy_signal = signal + noise
signal_max = np.max(np.abs(signal))
signal_normalized = signal / signal_max
noisy_signal_normalized = noisy_signal / signal_max

# 定义生成器（基于自编码器结构）
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv1d(1, 64, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2),
            nn.Conv1d(64, 32, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2),
            nn.Conv1d(32, 16, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2),
            nn.Conv1d(16, 8, kernel_size=5, padding=2),
            nn.ReLU(True),
            nn.MaxPool1d(2, stride=2)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose1d(8, 16, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.ReLU(True),
            nn.ConvTranspose1d(16, 32, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.ReLU(True),
            nn.ConvTranspose1d(32, 64, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.ReLU(True),
            nn.ConvTranspose1d(64, 1, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.Tanh()  # 输出层使用Tanh
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# 定义鉴别器
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=5, stride=2, padding=2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(16, 32, kernel_size=5, stride=2, padding=2),
            nn.BatchNorm1d(32),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(32, 64, kernel_size=5, stride=2, padding=2),
            nn.BatchNorm1d(64),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Flatten(),
            nn.Linear(64 * 128, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        output = self.model(x)
        return output

# 初始化模型、损失函数和优化器
generator = Generator()
discriminator = Discriminator()
g_optimizer = optim.Adam(generator.parameters(), lr=1e-4)
d_optimizer = optim.Adam(discriminator.parameters(), lr=1e-4)
adversarial_loss = nn.BCELoss()
reconstruction_loss = nn.MSELoss()

# 准备数据
X_train = torch.FloatTensor(noisy_signal_normalized).unsqueeze(0).unsqueeze(0)
Y_train = torch.FloatTensor(signal_normalized).unsqueeze(0).unsqueeze(0)

# 训练GAN
num_epochs = 500
for epoch in range(num_epochs):
    # 训练鉴别器
    real_loss = adversarial_loss(discriminator(Y_train), torch.ones_like(discriminator(Y_train)))
    fake_signal = generator(X_train)
    fake_loss = adversarial_loss(discriminator(fake_signal.detach()), torch.zeros_like(discriminator(fake_signal)))
    d_loss = (real_loss + fake_loss) / 2

    d_optimizer.zero_grad()
    d_loss.backward()
    d_optimizer.step()

    # 训练生成器
    g_loss = adversarial_loss(discriminator(fake_signal), torch.ones_like(discriminator(fake_signal))) + 100 * reconstruction_loss(fake_signal, Y_train)

    g_optimizer.zero_grad()
    g_loss.backward()
    g_optimizer.step()

    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch+1}/{num_epochs} D Loss: {d_loss.item():.4f} G Loss: {g_loss.item():.4f}")

# 测试生成器
generator.eval()
with torch.no_grad():
    filtered_signal_normalized = generator(X_train).squeeze(0).squeeze(0).numpy()
filtered_signal = filtered_signal_normalized * signal_max  # 恢复到原始幅度

# 计算性能指标
snr_noisy = compute_snr(noisy_signal, signal)
snr_filtered = compute_snr(filtered_signal, signal)
ncc = compute_ncc(filtered_signal, signal)
mse = reconstruction_loss(torch.FloatTensor(filtered_signal), torch.FloatTensor(signal)).item()
rmse = np.sqrt(mse)

print(f"SNR of Noisy Signal: {snr_noisy:.2f} dB")
print(f"SNR of Filtered Signal: {snr_filtered:.2f} dB")
print(f"NCC between Filtered and Original Signal: {ncc:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

# 绘制信号对比图
plt.figure(figsize=(14, 7))
plt.subplot(2,1,1)
plt.plot(t, signal, label='Original Signal', color='green')
plt.plot(t, noisy_signal, label='Noisy Signal', color='red')
plt.legend()

plt.subplot(2,1,2)
plt.plot(t, signal, label='Original Signal', color='green')
plt.plot(t, filtered_signal, label='Filtered Signal by GAN', color='blue')
plt.legend()
plt.show()
```

#  GAN+自编码器滤波（Liner）

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# 环境设置
plt.rcParams['font.sans-serif'].insert(0, 'SimHei')
plt.rcParams['axes.unicode_minus'] = False

# 生成数据
np.random.seed(0)
t = np.linspace(0, 1, 1024)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)
noise = np.random.normal(0, 0.5, signal.shape)
noisy_signal = signal + noise
signal_max = np.max(np.abs(signal))
signal_normalized = signal / signal_max
noisy_signal_normalized = noisy_signal / signal_max

# 定义生成器
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(1024, 512),
            nn.ReLU(True),
            nn.Linear(512, 256),
            nn.ReLU(True),
            nn.Linear(256, 128),
            nn.ReLU(True),
            nn.Linear(128, 256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, 1024),
            nn.Tanh()
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Flatten the input
        x = self.main(x)
        return x.view(-1, 1, 1024)  # Reshape to original shape

# 定义判别器
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(1024, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Flatten the input
        return self.main(x)

# 初始化模型、损失函数和优化器
generator = Generator()
discriminator = Discriminator()
g_optimizer = optim.Adam(generator.parameters(), lr=1e-4)
d_optimizer = optim.Adam(discriminator.parameters(), lr=1e-4)
adversarial_loss = nn.BCELoss()
reconstruction_loss = nn.MSELoss()

# 准备数据
X_train = torch.FloatTensor(noisy_signal_normalized).unsqueeze(0)
Y_train = torch.FloatTensor(signal_normalized).unsqueeze(0)

# 训练GAN
num_epochs = 250
for epoch in range(num_epochs):
    # 训练鉴别器
    real_loss = adversarial_loss(discriminator(Y_train), torch.ones_like(discriminator(Y_train)))
    fake_signal = generator(X_train)
    fake_loss = adversarial_loss(discriminator(fake_signal.detach()), torch.zeros_like(discriminator(fake_signal)))
    d_loss = (real_loss + fake_loss) / 2

    d_optimizer.zero_grad()
    d_loss.backward()
    d_optimizer.step()

    # 训练生成器
    g_loss = adversarial_loss(discriminator(fake_signal), torch.ones_like(discriminator(fake_signal))) + 100 * reconstruction_loss(fake_signal, Y_train)

    g_optimizer.zero_grad()
    g_loss.backward()
    g_optimizer.step()

    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch+1}/{num_epochs} D Loss: {d_loss.item():.4f} G Loss: {g_loss.item():.4f}")

# 测试生成器
generator.eval()
with torch.no_grad():
    filtered_signal_normalized = generator(X_train).squeeze(0).numpy()
filtered_signal = filtered_signal_normalized * signal_max

# 计算性能指标函数
def compute_snr(signal, reference):
    noise = signal - reference
    snr = 10 * np.log10(np.sum(reference**2) / np.sum(noise**2))
    return snr

def compute_ncc(signal, reference):
    # 确保输入是一维的
    signal = signal.flatten()
    reference = reference.flatten()
    ncc = np.correlate(signal, reference, mode='full')
    return ncc[ncc.size // 2] / (np.sqrt(np.sum(signal**2) * np.sum(reference**2)))

# Calculate performance metrics
# Ensure signal and filtered_signal are flattened 1D arrays
snr_noisy = compute_snr(noisy_signal.flatten(), signal.flatten())
snr_filtered = compute_snr(filtered_signal.flatten(), signal.flatten())
ncc = compute_ncc(filtered_signal.flatten(), signal.flatten())

# Compute the mean squared error
mse = reconstruction_loss(torch.tensor(filtered_signal.flatten()), torch.tensor(signal.flatten())).item()
# Compute the root mean squared error
rmse = np.sqrt(mse)

print(f"SNR of Noisy Signal: {snr_noisy:.2f} dB")
print(f"SNR of Filtered Signal: {snr_filtered:.2f} dB")
print(f"NCC between Filtered and Original Signal: {ncc:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

# Plot the comparison charts
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(t, signal.flatten(), label='Original Signal', color='green')     # Flatten signal here
plt.plot(t, noisy_signal.flatten(), label='Noisy Signal', color='red')    # Flatten noisy_signal here
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, signal.flatten(), label='Original Signal', color='green')     # Flatten signal here
plt.plot(t, filtered_signal.flatten(), label='Filtered Signal by GAN', color='blue') # Flatten filtered_signal here
plt.legend()

plt.show()
```
